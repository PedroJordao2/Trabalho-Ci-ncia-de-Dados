## Contexto

O desenvolvimento humano é um dos principais indicadores da qualidade de vida de uma população. No Brasil, esse desenvolvimento apresenta fortes desigualdades regionais e sociais, refletidas em diferenças significativas nos níveis de renda, educação, saúde e acesso a serviços básicos entre os municípios.

O **Índice de Desenvolvimento Humano Municipal (IDHM)** é amplamente utilizado para medir esse desenvolvimento, combinando três dimensões fundamentais:

- Educação  
- Renda  
- Longevidade  

Apesar de sua importância, o IDHM é um indicador agregado, o que dificulta a compreensão detalhada dos fatores que mais influenciam suas variações.

---

## Sobre os Dados

Para esta análise, foi utilizado o dataset do **Atlas do Desenvolvimento Humano (ADH)**, que reúne indicadores socioeconômicos e demográficos dos municípios brasileiros ao longo do tempo.

Esses dados permitem análises tanto **temporais** quanto **regionais**, possibilitando uma visão mais completa do desenvolvimento humano no país.

# Limpeza e Integração dos Dados

Após a análise inicial, identificamos a presença de valores nulos e a necessidade de integrar os datasets para enriquecer a análise.

Nesta etapa, iremos:

- Selecionar a chave de integração (`id_municipio`)
- Realizar o merge (união) entre os datasets
- Tratar valores ausentes quando necessário
- Preparar os dados para a análise exploratória (EDA)

Essa etapa é fundamental para garantir consistência e qualidade nos dados utilizados nas próximas análises.

```python
# Realizando a junção pelo id do município
df_merged = df.merge(municipios, on="id_municipio", how="left")

print("Dimensões após o merge:", df_merged.shape)
display(df_merged.head())
```

O merge foi realizado utilizando a chave `id_municipio`, permitindo integrar os indicadores socioeconômicos com as informações descritivas dos municípios. Como utilizamos um *left join*, garantimos que todos os registros do dataset principal foram mantidos, evitando perda de dados.

Após o merge, o dataset passou a ter **256 colunas**, indicando que as informações do dataset auxiliar foram incorporadas com sucesso ao dataset principal. O número de linhas permaneceu o mesmo (16.695), o que confirma que não houve perda de registros durante a integração.

Isso mostra que o merge foi realizado corretamente e que agora o dataset está mais completo, permitindo análises mais ricas ao combinar indicadores socioeconômicos com informações geográficas dos municípios.

## Verificação após o Merge

Após a integração dos dados, é importante verificar:

- Se houve perda de registros
- Se novas colunas foram adicionadas corretamente
- Se surgiram novos valores nulos

Essa validação garante que o merge foi realizado corretamente.

```python
# Verificando valores nulos após merge
display(df_merged.isnull().sum().sort_values(ascending=False).head(10))
```

Após o merge, observa-se que algumas colunas apresentam uma quantidade significativa de valores nulos, como `id_regiao_metropolitana` e `renda_media_ocupados`. Isso indica que essas informações não estão disponíveis para todos os municípios ou períodos analisados.

Além disso, várias variáveis com cerca de 5.565 valores nulos reforçam a ideia de que certos dados podem estar ausentes em determinados anos. Esses padrões serão considerados na etapa de limpeza, priorizando variáveis mais completas e relevantes para a análise.

## Tratamento de Valores Nulos

A análise mostrou que os valores nulos não estão distribuídos aleatoriamente, mas seguem padrões específicos:

- Algumas variáveis possuem muitos valores ausentes, indicando baixa disponibilidade de dados
- Outras apresentam ausência parcial, possivelmente relacionada a determinados anos ou regiões

Dessa forma, adotamos uma abordagem estratégica:

- Manter variáveis essenciais (como IDHM e indicadores principais)
- Remover colunas com muitos valores nulos e baixa relevância
- Preservar variáveis importantes mesmo com alguns dados ausentes

Essa estratégia garante equilíbrio entre qualidade dos dados e riqueza da análise.

```python
# 1. Definir colunas essenciais (não remover)
colunas_essenciais = [
    "idhm", "idhm_e", "idhm_l", "idhm_r",
    "renda_pc", "expectativa_vida", "indice_gini"
]

# 2. Remover colunas com muitos nulos (mais de 60%)
threshold = len(df_merged) * 0.4  # mantém colunas com pelo menos 40% preenchido

df_clean = df_merged.dropna(axis=1, thresh=threshold)

# 3. Garantir que colunas essenciais não foram removidas
for col in colunas_essenciais:
    if col not in df_clean.columns:
        df_clean[col] = df_merged[col]

print("Dimensões após limpeza:", df_clean.shape)
```

Após o processo de limpeza, o dataset passou a ter **253 colunas**, indicando que apenas algumas variáveis com alta quantidade de valores nulos foram removidas. O número de linhas foi mantido, garantindo que nenhuma observação foi perdida.

Isso mostra que a estratégia adotada foi equilibrada: reduzimos o impacto de dados incompletos sem comprometer a riqueza do dataset, preservando as variáveis mais relevantes para a análise exploratória.

## Ajustes Finais

Após a limpeza:

- Os dados estão mais consistentes
- Reduzimos o impacto de valores ausentes
- Mantivemos as variáveis mais relevantes para análise

Agora o dataset está preparado para a etapa de **Análise Exploratória de Dados (EDA)**.

```python
# Conferindo dataset final
display(df_clean.head())
```

## Análise Bivariada

A análise bivariada tem como objetivo investigar a relação entre duas variáveis, permitindo identificar padrões, correlações e possíveis associações entre os dados.

Nesta etapa, iremos analisar como diferentes fatores socioeconômicos se relacionam com o desenvolvimento humano (IDHM), buscando responder às perguntas norteadoras definidas anteriormente.

As principais relações analisadas serão:

- IDHM × Renda per capita  
- IDHM × Expectativa de vida  
- IDHM × Índice de Gini  

## Relação entre IDHM e Renda per Capita

A renda é um dos principais componentes do desenvolvimento humano. Espera-se que municípios com maior renda apresentem melhores níveis de IDHM.

O gráfico de dispersão permite visualizar essa relação e identificar possíveis padrões.

```python
sns.scatterplot(
    data=df_clean,
    x="renda_pc",
    y="idhm",
    alpha=0.5
)

plt.title("IDHM vs Renda per Capita")
plt.xlabel("Renda per Capita")
plt.ylabel("IDHM")
plt.show()
```

O gráfico mostra uma **relação positiva clara entre renda per capita e IDHM**, indicando que municípios com maior renda tendem a apresentar maiores níveis de desenvolvimento humano. No entanto, essa relação não é linear: observa-se que, à medida que a renda aumenta, os ganhos no IDHM passam a crescer mais lentamente.

Esse padrão sugere um efeito de **retornos decrescentes**, onde aumentos iniciais de renda têm grande impacto no desenvolvimento, mas, em níveis mais altos, esse impacto se reduz. Além disso, a dispersão dos pontos indica que a renda não é o único fator determinante, reforçando a importância de outras dimensões, como educação e saúde.

## Relação entre IDHM e Expectativa de Vida

A expectativa de vida está diretamente ligada à dimensão de longevidade do IDHM, sendo um indicador importante da qualidade de vida.

A análise dessa relação ajuda a entender como a saúde influencia o desenvolvimento humano.

```python
sns.scatterplot(
    data=df_clean,
    x="expectativa_vida",
    y="idhm",
    alpha=0.5
)

plt.title("IDHM vs Expectativa de Vida")
plt.xlabel("Expectativa de Vida")
plt.ylabel("IDHM")
plt.show()
```

O gráfico evidencia uma **forte relação positiva e praticamente linear entre expectativa de vida e IDHM**, indicando que municípios com maior longevidade tendem a apresentar níveis mais altos de desenvolvimento humano.

A baixa dispersão em torno da tendência sugere que essa é uma das relações mais consistentes do conjunto de dados, reforçando a importância da saúde e qualidade de vida como componentes fundamentais do desenvolvimento. Diferente da renda, essa relação se mostra mais estável, indicando que a expectativa de vida é um fator fortemente associado ao IDHM.

## Relação entre IDHM e Desigualdade (Gini)

O índice de Gini mede a desigualdade de renda. Espera-se que níveis mais altos de desigualdade estejam associados a menores níveis de desenvolvimento humano.

Essa análise permite verificar se essa relação se confirma nos dados.

```python
sns.scatterplot(
    data=df_clean,
    x="indice_gini",
    y="idhm",
    alpha=0.5
)

plt.title("IDHM vs Índice de Gini")
plt.xlabel("Índice de Gini")
plt.ylabel("IDHM")
plt.show()
```

A relação entre o índice de Gini e o IDHM não apresenta um padrão tão claro quanto as análises anteriores, indicando uma **correlação mais fraca ou menos direta** entre desigualdade de renda e desenvolvimento humano.

Embora seja possível observar uma leve tendência de que valores mais altos de desigualdade estejam associados a menores níveis de IDHM, a alta dispersão dos pontos sugere que outros fatores, como educação e saúde, podem ter um papel mais determinante. Isso reforça a ideia de que o desenvolvimento humano é um fenômeno multidimensional, não explicado por uma única variável.

## Correlação entre Variáveis

O mapa de calor (heatmap) permite visualizar a força e direção da relação entre múltiplas variáveis ao mesmo tempo.

Valores próximos de:
- **1** → correlação positiva forte  
- **-1** → correlação negativa forte  
- **0** → pouca ou nenhuma correlação  

```python
# Selecionando variáveis principais
cols = ["idhm", "renda_pc", "expectativa_vida", "indice_gini"]

corr = df_clean[cols].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlação")
plt.show()
```

A matriz de correlação confirma os padrões observados anteriormente. O IDHM apresenta **forte correlação positiva com a expectativa de vida (0.92)** e com a renda per capita (0.83), indicando que essas variáveis são fortemente associadas ao desenvolvimento humano.

Por outro lado, o índice de Gini apresenta **correlação negativa fraca** com o IDHM (-0.22), sugerindo que a desigualdade tem influência, mas menor do que outros fatores. Esses resultados reforçam que o desenvolvimento humano está mais diretamente ligado a condições de vida e renda do que à desigualdade isoladamente.

# Evolução do Desenvolvimento Humano ao Longo do Tempo
