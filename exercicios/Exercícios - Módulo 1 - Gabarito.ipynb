Exercício 2 — Lista e média

Crie uma lista com 5 números e calcule a média.

```python
numeros = [10, 7, 8, 9, 6]

media = sum(numeros) / len(numeros)

print("Lista:", numeros)
print("Média:", media)
```

Exercício 4 — Média entre dois números

Crie um sistema que recebe dois números diferentes e exibe a média entre eles.

```python
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))

media = (num1 + num2) / 2

print("A média é:", media)
```

Exercício 6 — Sistema simples de notas

Simule um sistema simples de notas:

- Peça 3 notas
- Calcule a média
- Informe se o aluno foi aprovado (média ≥ 6)

```python
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))

media = (nota1 + nota2 + nota3) / 3

print("Média do aluno:", media)

if media >= 6:
    print("Aluno APROVADO")
else:
    print("Aluno REPROVADO")
```
