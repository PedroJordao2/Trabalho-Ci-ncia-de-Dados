```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# Ajuste o caminho do arquivo, se necessário
df = pd.read_csv("../datasets/brazilian_cities.csv")

df.head()
```

## 1. Distribuição etária da população

Crie um gráfico de barras mostrando a **distribuição média das faixas etárias**:

- `IBGE_1`
- `IBGE_1-4`
- `IBGE_5-9`
- `IBGE_10-14`
- `IBGE_15-59`
- `IBGE_60+`

**Objetivo:** entender o perfil etário médio dos municípios.

### Gabarito

```python
# Seleciona as colunas de faixas etárias que existem no dataset
colunas_idade = ["IBGE_1", "IBGE_1-4", "IBGE_5-9", "IBGE_10-14", "IBGE_15-59", "IBGE_60+"]
colunas_idade = [c for c in colunas_idade if c in df.columns]

# Calcula a média de cada faixa etária
media_idade = df[colunas_idade].mean()

# Cria o gráfico
plt.figure(figsize=(10, 6))
media_idade.plot(kind="bar")

plt.title("População média por faixa etária nos municípios brasileiros")
plt.xlabel("Faixa etária (anos)")
plt.ylabel("Número médio de habitantes por município")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 2. Relação entre área e população

Crie um **scatterplot** entre:

- `AREA`
- `ESTIMATED_POP`

**Pergunta:** municípios maiores em área tendem a ser mais populosos?

### Gabarito

```python
df_plot = df.dropna(subset=["AREA", "ESTIMATED_POP"])

plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df_plot,
    x="AREA",
    y="ESTIMATED_POP",
    alpha=0.6
)

plt.ticklabel_format(style='plain', axis='y')
plt.ticklabel_format(style='plain', axis='x')

plt.title("Relação entre área territorial e população estimada")
plt.xlabel("Área (km²)")
plt.ylabel("População estimada")

plt.tight_layout()
plt.show()
```

## 3. Densidade demográfica

1. Crie uma nova coluna:

`DENSIDADE = ESTIMATED_POP / AREA`

2. Faça um **histograma** dessa nova variável.

**Objetivo:** entender a concentração populacional.

### Gabarito

```python
# Cria uma cópia removendo valores nulos
df_plot = df.dropna(subset=["AREA", "ESTIMATED_POP"]).copy()

# Cria a variável densidade demográfica
df_plot["DENSIDADE"] = df_plot["ESTIMATED_POP"] / df_plot["AREA"]

# Cria o histograma
plt.figure(figsize=(10, 6))
sns.histplot(df_plot["DENSIDADE"], bins=30, kde=True)

plt.title("Distribuição da densidade demográfica dos municípios brasileiros")
plt.xlabel("Densidade demográfica (habitantes por km²)")
plt.ylabel("Número de municípios")
plt.tight_layout()
plt.show()
```

## 4. População urbana vs IDHM

Utilize:

- `RURAL_URBAN`
- `IDHM`

Crie um boxplot comparando o IDHM entre áreas urbanas e rurais

**Pergunta:** Existe diferença no nível de desenvolvimento entre áreas urbanas e rurais?

### Gabarito

## 5. Empresas por setor (composição)

Escolha os setores:

- `COMP_A`
- `COMP_G`
- `COMP_K`
- `COMP_M`

Crie um gráfico de **barras empilhadas** mostrando a composição média desses setores por estado.

### Gabarito

```python
# Seleciona as colunas desejadas
cols = ["COMP_A", "COMP_G", "COMP_K", "COMP_M"]
cols = [c for c in cols if c in df.columns]

# Remove linhas em que todos os setores estão nulos
df_plot = df.dropna(subset=cols, how="all")

# Calcula a média por estado
setores = df_plot.groupby("STATE")[cols].mean().head(10)

# Converte para percentual para destacar composição
setores_pct = setores.div(setores.sum(axis=1), axis=0) * 100

# Cria o gráfico empilhado
setores_pct.plot(kind="bar", stacked=True, figsize=(12, 6))

plt.title("Composição de empresas por setor")
plt.xlabel("Estado")
plt.ylabel("Percentual (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
