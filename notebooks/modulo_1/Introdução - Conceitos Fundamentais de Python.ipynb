# 4️⃣ Estruturas Condicionais

```python
idade = 18

if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
```

```python
idade = 20
tem_carteira = True

if idade >= 18 and tem_carteira:
    print("Pode dirigir")
```

# 5️⃣ Estruturas de Repetição

```python
for i in range(5):
    print(i)
```

```python
contador = 0

while contador < 5:
    print(contador)
    contador += 1
```
